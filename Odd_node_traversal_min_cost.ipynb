```python
def shortest_path_odd_node_traversal(start,finish):
    return(nx.dijkstra_path(g1,start,finish),nx.dijkstra_path_length(g1,strt,finish))
```

```python
shortest_path=[]
try:
    shortest_path,cost=shortest_path_odd_node_traversal(strt,dest)
    print("Shortest path between {} and {} by odd-node traversal :\n".format(strt,dest),shortest_path)
    print("Cost of Traversal:",cost)
except nx.NetworkXNoPath:
    print("No path found between {} and {}".format(strt,dest))
```
